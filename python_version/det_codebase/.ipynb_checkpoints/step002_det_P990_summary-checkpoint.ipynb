import packages
------------------

Some packages are installed automatically if you use Anaconda. As pytorch is used here, you are expected to install that in your machine. 

```python
from __future__ import print_function, division
import os, random, time, copy
import numpy as np
import os.path as path
import scipy.io as sio
from scipy import misc
from scipy import ndimage, signal
import scipy
import pickle
import math
import matplotlib.pyplot as plt
import PIL.Image
```

```python
save_to_folder = './result' # experiment directory, used for reading the init model


filename = path.join(save_to_folder, 'step002_det_P001_segVoting_v1.pkl')    
with open(filename, 'rb') as handle:
    result_model1 = pickle.load(handle)
    
filename = path.join(save_to_folder, 'step002_det_P002_segDistTransform_v1.pkl')    
with open(filename, 'rb') as handle:
    result_model2 = pickle.load(handle)
    
filename = path.join(save_to_folder, 'step002_det_P003_keyRadius_v3_distTransform_noSeg.pkl')    
with open(filename, 'rb') as handle:
    result_model3 = pickle.load(handle)
    
```

draw curve
-----

```python
iou_model1 = np.asarray(result_model1['IoUList'], dtype=np.float32)
iou_model2 = np.asarray(result_model2['IoUList'], dtype=np.float32)
iou_model3 = np.asarray(result_model3['IoUList'], dtype=np.float32)

recallList_model1 = []
recallList_model2 = []
recallList_model3 = []
iouThreshList = (np.arange(9).astype(np.float32)+1)/10
for i in range(len(iouThreshList)):
    tmp = (iou_model1>iouThreshList[i]).astype(np.float32)    
    recallList_model1 += [tmp.mean()]
    
    tmp = (iou_model2>iouThreshList[i]).astype(np.float32)    
    recallList_model2 += [tmp.mean()]
    
    tmp = (iou_model3>iouThreshList[i]).astype(np.float32)    
    recallList_model3 += [tmp.mean()]
    
recallList_model1 = np.asarray(recallList_model1, dtype=np.float32)
recallList_model2 = np.asarray(recallList_model2, dtype=np.float32)
recallList_model3 = np.asarray(recallList_model3, dtype=np.float32)
```

```python
fig, ax1 = plt.subplots()


legendList = ['model-1 avgRecall:{:.3f}: centerVoting'.format(recallList_model1.mean()), 
              'model-2 avgRecall:{:.3f}: Distance Transform'.format(recallList_model2.mean()),
              'model-3 avgRecall:{:.3f}: center+radius prediction'.format(recallList_model3.mean())]

ax1.set_xlabel(r'iou as threshold', size=15)
ax1.set_ylabel('Recall', size=15)

ax1.plot(iouThreshList, recallList_model1, color='black', linewidth=3, linestyle='-', label=legendList[0])
ax1.plot(iouThreshList, recallList_model2, color='blue',linewidth=3, linestyle=':', label=legendList[1])
ax1.plot(iouThreshList, recallList_model3, color='red',  linewidth=3, linestyle='-.', label=legendList[2])
ax1.grid()
ax1.tick_params(axis='both', which='major', labelsize=15)

box = ax1.get_position()
ax1.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=15)
```
